```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
df = pd.read_csv("Resources/cities.csv")
print(df.to_html())
```

```python
df
```

```python
plt.scatter(df["Lng"],df["Lat"],c=df["Humidity"],cmap=cm.get_cmap("coolwarm"))
plt.title("Humidity vs Location")
plt.savefig("Plot1")
```

```python
plt.scatter(df["Lng"],df["Lat"],c=df["Max Temp"],cmap=cm.get_cmap("coolwarm"))
plt.title("Max Temperature vs Location")
plt.savefig("Plot2")
```

```python
plt.scatter(df["Lng"],df["Lat"],c=df["Wind Speed"],cmap=cm.get_cmap("coolwarm"))
plt.title("Wind Speed vs Location")
plt.savefig("Plot3")
```

```python
plt.scatter(df["Lng"],df["Lat"],c=df["Cloudiness"],cmap=cm.get_cmap("coolwarm"))
plt.title("Cloudiness vs Location")
plt.savefig("Plot4")
```
